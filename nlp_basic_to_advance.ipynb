```python

def extract_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    full_name = get_pattern_match('Born(.*)\n', text)
    birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match('\(age.*\n(.*)', text)
    return {
        'age': int(age),
        'name': full_name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip()
    }
```

```python
text = """Latha is very multi talented girl.She is good at many skills like dancing, running, singing, playing.She also likes eating Pav Bhagi. she has a
habit of fishing and swimming too.Besides all this, she is a wonderful at cooking too.
"""

#using stemming in nltk

#step1: Word tokenizing
all_word_tokens = nltk.word_tokenize(text)


#step2: getting the base form for each token using stemmer
all_base_words = []

for token in all_word_tokens:
  base_form = stemmer.stem(token)
  all_base_words.append(base_form)


#step3: joining all words in a list into string using 'join()'
final_base_text = ' '.join(all_base_words)
print(final_base_text)
```

```python
#using lemmatisation in spacy


#step1: Creating the object for the given text
doc = nlp(text)
all_base_words = []

#step2: getting the base form for each token using spacy 'lemma_'
for token in doc:
  base_word =  token.lemma_
  all_base_words.append(base_word)


#step3: joining all words in a list into string using 'join()'
final_base_text = ' '.join(all_base_words)
print(final_base_text)
```

```python
import spacy
from spacy.matcher import Matcher

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Initialize the Matcher with the loaded nlp object
matcher = Matcher(nlp.vocab)

# Define the patterns
pattern1 = [{"LOWER": "solarpower"}]
pattern2 = [{"LOWER": "solar"}, {"LOWER": "power"}]
pattern3 = [{"LOWER": "solarpower"}, {"IS_PUNCT": True}, {"LOWER": "solar"}]

# Add patterns to the Matcher
patterns = [pattern1, pattern2, pattern3]
matcher.add("SolarPower", patterns)

# Example text to test the matcher
text = "Solarpower is the conversion of sunlight into electricity, either directly using photovoltaics, or indirectly using concentrated solar power."

# Process the text with spaCy
doc = nlp(text)

# Apply the Matcher to the processed text
matches = matcher(doc)
print(matches)

# Iterate over the matches and print the matched text
for match_id, start, end, in matches:
    matched_span = doc[start:end]
    print(matched_span.text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
print("Pipeline:", nlp.pipe_names)
doc = nlp("I was reading the paper.")
token = doc[0]  # 'I'
print(token.morph)  # 'Case=Nom|Number=Sing|Person=1|PronType=Prs'
print(token.morph.get("PronType"))  # ['Prs']
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Print the names of components in the spaCy processing pipeline
print("Pipeline:", nlp.pipe_names)

# Process the input text
doc = nlp("I was reading the paper.")

# Access the first token in the processed document
token = doc[0]  # 'I'

# Access the morphological information of the token
print(token.morph)  # 'Case=Nom|Number=Sing|Person=1|PronType=Prs'

# Access a specific morphological feature of the token ('PronType')
print(token.morph.get("Paper"))  # ['Prs']
```

```python
import spacy
from spacy.tokens import Span

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Process the input text
doc = nlp("I was reading the paper.")

# Define the label for the new entity
org = doc.vocab.strings[u"PERSON"]

# Create a new Span representing the entity
new_ent = Span(doc, 0, 1, label=org)

# Add the new entity to the existing entities in the document
doc.ents = list(doc.ents) + [new_ent]

# Print the entities in the document after adding the new entity
print([(ent.text, ent.label_) for ent in doc.ents])
```

```python
from spacy.language import Language
import spacy

text = "this is a sentence...hello...and another sentence."

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
print("Before:", [sent.text for sent in doc.sents])

@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text == "...":
            doc[token.i + 1].is_sent_start = True
    return doc

nlp.add_pipe("set_custom_boundaries", before="parser")
doc = nlp(text)
print("After:", [sent.text for sent in doc.sents])
```

# n_gram

```python
def N_Grams(text, n):
    token = text.split()
    ngrams = []  # Initialize the list before appending to it
    for i in range(len(token)-n+1):
        temp = [token[j] for j in range(i, i+n)]
        ngrams.append(" ".join(temp))
    return ngrams
```

```python
text1 = "the quick brown fox jump the lazy dog"
```

```python
N_Grams(text1,1)
```

```python
N_Grams(text1,2)
```
